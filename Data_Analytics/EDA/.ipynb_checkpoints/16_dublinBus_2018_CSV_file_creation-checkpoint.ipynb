## Merge [monthName]_2018.csv  with trips_2018.csv
* rt_leavetimes_DB_2018.txt is of size ~ 11 GB. It is memory expensive to read and anlyze
* Hence it is devided into 1 GB chunks -- resulting into 11 subfiles
* Further, these **1 GB subfiles files are succesively processed to generate seperate files named <month_year.csv> of data for each month**
* Based on priliminary EDA, only following features are preserved while converting trips file into csv<br>
 [DAYOFSERVICE, TRIPID, PROGRNUMBER, STOPPOINTID, PLANNEDTIME_ARR, PLANNEDTIME_DEP, ACTUALTIME_ARR, ACTUALTIME_DEP]

### read rt_trips_DB_2018.csv
**rt_trips_DB_2018.txt consists data for ENTIRE 2018**
* Based on priliminary EDA, only following features are preserved while converting trips file into csv<br>
 [DAYOFSERVICE, TRIPID, LINEID, ROUTEID, DIRECTION, PLANNEDTIME_ARR, PLANNEDTIME_DEP, ACTUALTIME_ARR, ACTUALTIME_DEP ]

```python
# Function to read monthly data from csv and return a processed dataframe
def read_trips_data(file_trips):
    # read file
    df_trips =  pd.read_csv(file_trips,skip_blank_lines=True,index_col=False)
    
    # drop unwanted features [PLANNEDTIME_ARR, PLANNEDTIME_DEP, ACTUALTIME_ARR, ACTUALTIME_DEP]
    df_trips = df_trips.drop(labels = ['PLANNEDTIME_ARR','PLANNEDTIME_DEP', 'ACTUALTIME_ARR','ACTUALTIME_DEP']
                             ,axis=1
                            ).drop_duplicates()

    # typecast to suitable datatypes
    df_trips['DAYOFSERVICE'] = pd.to_datetime(df_trips['DAYOFSERVICE'])
    df_trips['TRIPID'] = pd.to_numeric(df_trips['TRIPID'], downcast='integer', errors='coerce')
    df_trips['LINEID'] = df_trips['LINEID'].astype('str')
    df_trips['ROUTEID'] = df_trips['ROUTEID'].astype('str')
    df_trips['DIRECTION'] = pd.to_numeric(df_trips['DIRECTION'], downcast='integer', errors='coerce')

    gc.collect()
    
    return df_trips
```

<br>

### read [month]_2018.csv & Filter data for presently served DUBLIN BUS STOPS
**Historic data consists many STOPPOINTIDs which no longer exist**  
* **Hence; another python script is written which parses stopID.csv and makes querry to RTPI web API** 
* **It generates a list of presently served Dublin Bus STOPPOINTIDs**  
* **Entries other than these presently served bus stops are ommitted**  

```python
# Function to read monthly data from csv and return a processed dataframe
def read_monthly_data(file_month, file_stops):
    # read file
    df_month =  pd.read_csv(file_month,skip_blank_lines=True,index_col=False)

    # drop unwanted feature [VEHICLEID]
    df_month.drop('VEHICLEID',axis=1,inplace=True)

    # typecast to suitable datatypes
    df_month['DAYOFSERVICE'] = pd.to_datetime(df_month['DAYOFSERVICE'])
    df_month['TRIPID'] = pd.to_numeric(df_month['TRIPID'], downcast='integer', errors='coerce')
    df_month['PROGRNUMBER'] = pd.to_numeric(df_month['PROGRNUMBER'], downcast='integer', errors='coerce')
    df_month['STOPPOINTID'] = pd.to_numeric(df_month['STOPPOINTID'], downcast='integer', errors='coerce')
    df_month['PLANNEDTIME_ARR'] = pd.to_numeric(df_month['PLANNEDTIME_ARR'], downcast='integer', errors='coerce')
    df_month['PLANNEDTIME_DEP'] = pd.to_numeric(df_month['PLANNEDTIME_DEP'], downcast='integer', errors='coerce')
    df_month['ACTUALTIME_ARR'] = pd.to_numeric(df_month['ACTUALTIME_ARR'], downcast='integer', errors='coerce')
    df_month['ACTUALTIME_DEP'] = pd.to_numeric(df_month['ACTUALTIME_DEP'], downcast='integer', errors='coerce')

    # filter data only for presently served bus stops
    df_presently_served_stops = pd.read_csv(file_stops)
    list_presently_served_stops = df_presently_served_stops["STOPPOINTID"].unique()
    df_month = df_month.loc[df_month['STOPPOINTID'].isin(list_presently_served_stops)]
    
    gc.collect()
    
    return df_month
```

### Support functions for file parsing

```python
# Return list of files with provided suffix in argument directory 
def find_files_extension(path_dir, suffix="2018.csv"):
    filenames = os.listdir(path_dir)
    
    gc.collect()
    
    return [filename for filename in filenames if filename.endswith(suffix)]
```

<br>

### Merge df_month with df_trips on 'DAYOFSERVICE','TRIPID'
**Make merged dataframe using all CSV files (i.e. month_year.csv files) in folder successively**

```python
def merge_leavetimes_trips(leavetimes_path, trips_path, file_stops):

    # Instantiate emtry dataframe for 2018 dublin bus data
    df_dbus_2018 = pd.DataFrame()

    # Read trips data into dataframe
    files = find_files_extension(trips_path,  suffix="2018.csv")
    for file in files:
        print("File processing:",trips_path+file)
        df_trips = read_trips_data(trips_path+str(file))

    # Read leavetimes monthly data into dataframe and merge it with leavetimes
    files = find_files_extension(leavetimes_path,  suffix="2018.csv")
    for file in files:
        print("File processing:",leavetimes_path+file)
        df_month = read_monthly_data(leavetimes_path+str(file), file_stops)

        # merge month data with trips data 
        df_month_trips = df_month.merge(df_trips,on=['DAYOFSERVICE','TRIPID']).drop_duplicates()
        del df_month
        
        # concat data to dublin bus 2018 merged dataframe
        df_dbus_2018 = pd.concat([df_dbus_2018,df_month_trips]).reset_index(drop=True)
        del df_month_trips
        
        print("shape of dataframe after merging",file,":", df_dbus_2018.shape)
        
    df_dbus_2018 = df_dbus_2018.drop_duplicates() 

    gc.collect()

    return df_dbus_2018
```

## Drop insignificant ROUTEIDS
**From prior EDA; We have obtained source of truth stating that:**
* **For a LINEID; ROUTEIDs and TRIPIDs used for DIFFERENT DIRECTION are mutually exclusive**
* **EVEN FOR ROUTEIDs HAVING SAME DIRECTION; STOPPOINTIDs VISITED ARE DIFFERENT**
* **ROUTEIDs ARE NOT KNOWN TO USER & ROUTEIDs ARE NOT TIMEBOUND i.e. timetable DOES NOT exists to follow a ROUTEID**
* **majority of data for a LINEID is tied with a particular ROUTEID**

**Hence, we can safely drop data for insignificant ROUTEIDs** 

```python
## Uncomment and Run to start operation from current section

df_dbus_2018 = pd.read_csv("../DB/ML/leavetimes_trips_MERGED_2018.csv")

df_dbus_2018['DAYOFSERVICE'] = pd.to_datetime(df_dbus_2018['DAYOFSERVICE'])
df_dbus_2018['TRIPID'] = pd.to_numeric(df_dbus_2018['TRIPID'], downcast='integer', errors='coerce')
df_dbus_2018['PROGRNUMBER'] = pd.to_numeric(df_dbus_2018['PROGRNUMBER'], downcast='integer', errors='coerce')
df_dbus_2018['STOPPOINTID'] = pd.to_numeric(df_dbus_2018['STOPPOINTID'], downcast='integer', errors='coerce')
df_dbus_2018['PLANNEDTIME_ARR'] = pd.to_numeric(df_dbus_2018['PLANNEDTIME_ARR'], downcast='integer', errors='coerce')
df_dbus_2018['PLANNEDTIME_DEP'] = pd.to_numeric(df_dbus_2018['PLANNEDTIME_DEP'], downcast='integer', errors='coerce')
df_dbus_2018['ACTUALTIME_ARR'] = pd.to_numeric(df_dbus_2018['ACTUALTIME_ARR'], downcast='integer', errors='coerce')
df_dbus_2018['ACTUALTIME_DEP'] = pd.to_numeric(df_dbus_2018['ACTUALTIME_DEP'], downcast='integer', errors='coerce')
df_dbus_2018['LINEID'] = df_dbus_2018['LINEID'].astype('str')
df_dbus_2018['ROUTEID'] = df_dbus_2018['ROUTEID'].astype('str')
df_dbus_2018['DIRECTION'] = pd.to_numeric(df_dbus_2018['DIRECTION'], downcast='integer', errors='coerce')
```

### Check distribution for each ROUTEID of each LINEID
**From above evaluation; we do not find definite titmetable is observed to operate a specific ROUTEID**
* **Hence, we need to consider data from specific ROUTEID only and drop the rest**
* **Before taking the step, we should ensure that:**
    * **There exists a dominant ROUTEID for either direction of LINEID**

```python
# plot counts of entries for ROUTEID of a particular LINEID

df_dominant_route = pd.DataFrame(columns=["DIRECTION","LINEID","ROUTEID"])

with PdfPages('../Reports/LINEID_Route_counts_JAN_data.pdf') as pp:

    # fetch list of LINEIDs operated
    list_LINEID = pd.read_csv("../DB/LINEID_2018.csv")['LINEID'].tolist()
    # for Line in list(df_dbus_2018.LINEID.unique()):
    for Line in list_LINEID:  
        # dataframe dominant_route is used to record dominant route for each LINEID
        df_line = df_dbus_2018.loc[df_dbus_2018["LINEID"] == Line]
        try:
            dom_route_D1 = df_line[(df_line["DIRECTION"] == 1)]["ROUTEID"].value_counts(sort=True).keys()[0]
            df_dominant_route.loc[len(df_dominant_route)] = [1,Line,dom_route_D1]
        except:
            print("error in line",Line,"direction 1")
        
        try:
            dom_route_D2 = df_line[(df_line["DIRECTION"] == 2)]["ROUTEID"].value_counts(sort=True).keys()[0]            
            df_dominant_route.loc[len(df_dominant_route)] = [2,Line,dom_route_D2]
        except:
            print("error in line",Line,"direction 2")
            
        plt.figure()
        sns.catplot(x="ROUTEID",col="DIRECTION",kind="count",palette=sns.color_palette(['orange', 'black']), \
                data=df_line)
        pp.savefig()
```

#### ROUND-OFF [FLOOR] PLANNEDTIME_ARR TO MINUTE RESOLUTION
* **PLANNEDTIME_ARR are in seconds resolution**
* **With these resolutions; earlier ML model gave RMSE of aproximately 300 seconds**
* **Flooring seconds data to minutes data would add worst case input error of 1 minute**
* **BUT AS OBSERVED BELLOW; IT REDUCES CARDINALITY OF FETURE SIGNIFICANTLY**

FOR JANUARY 2018 DATA:

|FEATURE|COUNT of unique entries(seconds)|COUNT of unique entries (minutes)|

|PLANNEDTIME_ARR  |67749  |1215  |


```python
df_weather_leavetimes_trips_arr_MINUTES['PLANNEDTIME_ARR'] = df_weather_leavetimes_trips_arr_MINUTES['PLANNEDTIME_ARR'] - df_weather_leavetimes_trips_arr_MINUTES['PLANNEDTIME_ARR'].mod(60)
```

### Generate LINES_PROGR.CSV

```python
## Uncomment and Run to start operation from current section

# df_STOPS_LINES_PROGR_DIR = pd.read_csv("../DB/ML/leavetimes_trips_arr_TIMETABLE.csv")

# df_STOPS_LINES_PROGR_DIR = df_STOPS_LINES_PROGR_DIR[['LINEID', 'DIRECTION','PROGRNUMBER']].drop_duplicates()

# df_STOPS_LINES_PROGR_DIR['LINEID'] = df_STOPS_LINES_PROGR_DIR['LINEID'].astype('category')
# df_STOPS_LINES_PROGR_DIR['DIRECTION'] = df_STOPS_LINES_PROGR_DIR['DIRECTION'].astype('category')
# df_STOPS_LINES_PROGR_DIR['PROGRNUMBER'] = df_STOPS_LINES_PROGR_DIR['PROGRNUMBER'].astype('category')
```

```python
# Define schema for target Dataframe
df_LINES_PROGR = pd.DataFrame(columns=['LINEID', 'DIRECTION','FIRST_PROGRNUMBER','LAST_PROGRNUMBER'])

# Take slice of df_STOPS_LINES_PROGR_DIR
df_LINES_PROGR_DIR = df_STOPS_LINES_PROGR_DIR[['LINEID', 'DIRECTION','PROGRNUMBER']].copy()

# group by Line and direction
group_LINES_PROGR_DIR = df_LINES_PROGR_DIR.groupby(['LINEID', 'DIRECTION'])['PROGRNUMBER']

for (line,direction),List_ProgrNum in group_LINES_PROGR_DIR:
    list_PROGR = sorted(List_ProgrNum.to_list())
    df_LINES_PROGR.loc[len(df_LINES_PROGR)] = [line,direction,list_PROGR[0],list_PROGR[-1]]

    # Generate data entry to meta  table
    df_LINES_PROGR.loc[len(df_LINES_PROGR)] = [line,direction,list_PROGR[0],list_PROGR[-1]]

df_LINES_PROGR.drop_duplicates(inplace = True)
```
